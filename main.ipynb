```python
def funcao_recomendacao(notas):
    media = notas.nota.mean()
    desvio_padrao = notas.nota.std()
    nota = {}
    for i in notas.index:
        nota[i] = notas.nota[i]
        if notas.nota[i] < (media-3*desvio_padrao):
            print("Abaixo da média - 3 desvios: \n{}:{}\n".format(i, notas.nota[i]))
        elif notas.nota[i] < (media-2*desvio_padrao):
            print("Abaixo da média - 2 desvios: \n{}:{}\n".format(i, notas.nota[i]))
        elif notas.nota[i] < (media-1*desvio_padrao):
            print("Abaixo da média - 1 desvio: \n{}:{}\n".format(i, notas.nota[i]))
        else:
            print("Acima da média: \n{}:{}\n".format(i, notas.nota[i]))
        
    #return nota, media, desvio_padrao
```
